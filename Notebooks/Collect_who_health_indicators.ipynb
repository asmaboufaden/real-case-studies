```python
"""
📊 This notebook extracts health indicators from the World Bank API — including Life Expectancy, Health Expenditure (% of GDP), and 
Maternal Mortality — for North African countries (Tunisia, Algeria, Morocco, Egypt, Libya). It cleans and filters the dataset for the years 2000–2023,
providing a ready-to-use foundation for evidence-based analysis, policy evaluation, and impactful visualizations.
"""
# Install required libraries
!pip install wbdata pandas

import wbdata
import pandas as pd
from datetime import datetime

# Define target countries (North Africa + Egypt + Libya)
countries = ['TUN', 'DZA', 'MAR', 'EGY', 'LBY']

# Define WHO/World Bank indicators
# (all available directly via the World Bank API)
indicators = {
    'SP.DYN.LE00.IN': 'Life_Expectancy',           # Life expectancy at birth
    'SH.XPD.CHEX.GD.ZS': 'Health_Expenditure_GDP', # Health expenditure (% of GDP)
    'SH.STA.MMRT': 'Maternal_Mortality',           # Maternal mortality ratio
   # 'SH.DYN.MORT': 'Child_Mortality',             # Child mortality (optional)
   # 'SH.MED.BEDS.ZS': 'Hospital_Beds',            # Hospital beds per 1,000 people (optional)
}

# Download directly in Python — no need to use the website
data = wbdata.get_dataframe(
    indicators,
    country=countries,
    # data_date=(datetime(2000, 1, 1), datetime(2023, 1, 1)) # not supported in get_dataframe
)

# Reset index to turn 'date' and 'country' into columns
data = data.reset_index()

# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter between 2000 and 2023
data = data[(data['date'].dt.year >= 2000) & (data['date'].dt.year <= 2023)]

# Export cleaned dataset to CSV for further analysis or visualization
data.to_csv('who_health_indicators.csv', index=False)

# Print dataset shape and first rows for verification
print("Downloaded dataset shape:", data.shape)
print(data.head())
```
