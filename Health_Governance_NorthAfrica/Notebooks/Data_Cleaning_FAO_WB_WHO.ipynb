```python

import pandas as pd
import numpy as np

# 1. load data
fao_data = pd.read_csv('faostat_food_security.csv') # https://www.fao.org/faostat/en  - "Food Security and Nutrition"
wb_data = pd.read_csv('worldbank_governance.csv') # https://databank.worldbank.org/source/worldwide-governance-indicators - World Bank Governance Indicators
who_data = pd.read_csv('who_health_indicators.csv') # https://www.who.int/data/gho/data/indicators - WHO Health Indicators

# 2. Inspect data FAO
fao_data.head(5)

```

```python
print(fao_data.info())
```

```python
print(fao_data.isnull().sum())  # missing values
```

```python
# 3. Cleaning
# Rename columns
fao_data.rename(columns={'Area': 'Country'}, inplace=True)
# Remove columns not useful in our analysis
fao_data = fao_data.drop(['Domain Code', 'Area Code (M49)', 'Element Code', 'Element', 'Flag', 'Note', 'Item Code', 'Flag Description', 'Year Code'], axis=1)

# Convert values to string first (if mixed types)
fao_data['Value'] = fao_data['Value'].astype(str)

# Replace "<2.5" with 2.49
fao_data['Value'] = fao_data['Value'].replace('<2.5', 2.49).astype(float)
```

```python
fao_data.tail(5)
```

```python
# Since the World Bank governance file is annual, the most practical and consistent choice is to take the middle year of the FAO interval,
# because it represents the central point of the average.

# Function to convert interval (e.g., "2000-2002") into middle year (2001)
def convert_interval_to_middle(year_str):
    if "-" in year_str:
        start, end = year_str.split("-")
        start, end = int(start), int(end)
        middle = (start + end) // 2
        return middle
    else:
        return int(year_str)

# Apply conversion
fao_data["Year"] = fao_data["Year"].astype(str).apply(convert_interval_to_middle)

# Check result
fao_data.head(15)
```

```python
# renaming the FAO “Item” column values into short, clear English names 

item_map = {
    'Prevalence of undernourishment (percent) (3-year average)': 'Undernourishment_pct',
    'Per capita food supply variability (kcal/cap/day)': 'FoodSupply_variability',
    'Percentage of population using safely managed drinking water services (percent)': 'SafeWater_pct'
}
fao_data['Item_short'] = fao_data['Item'].map(item_map)
fao_pivot = fao_data.pivot_table(
    index=['Country', 'Year'],
    columns='Item_short',
    values='Value',
    aggfunc='first'
).reset_index()
fao_pivot.columns.name = None

fao_pivot.head()
```

```python
print(fao_pivot.isnull().sum())  # missing values
```

```python
print("=== DIAGNOSTIC BEFORE IMPUTATION ===")
for col in ['Undernourishment_pct', 'SafeWater_pct', 'FoodSupply_variability']:
    print(f"\n{col} — missing values per country:")
    print(fao_pivot.groupby('Country')[col].apply(lambda x: x.isnull().sum()))
```

### ⚠️ Methodological Note — SafeWater_pct

Safe Water Access data (FAO) is **not available** for Egypt and Libya
for the period 2002–2023. These missing values have been **intentionally
retained as NaN** rather than imputed, as their absence reflects a
genuine data gap rather than a collection error.

Analyses involving this indicator are therefore **limited to Tunisia,
Algeria, and Morocco**.

This approach follows standard M&E practice:
*"The absence of data is itself an information."*

```python
# ============================================================
# STEP 1 — Undernourishment_pct
# Linear interpolation by country (all countries combined)
# ============================================================
fao_pivot['Undernourishment_pct'] = fao_pivot.groupby('Country')['Undernourishment_pct'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)
print("\n✅ Undernourishment_pct — linear interpolation applied")
```

```python
# ============================================================
# STEP 2 — SafeWater_pct
# Linear interpolation ONLY for Tunisia, Algeria, Morocco
# Egypt and Libya → NaN intentionally preserved
# ============================================================
countries_with_data = ['Tunisia', 'Algeria', 'Morocco']

def impute_safewater(group):
    country = group['Country'].iloc[0]
    if country in countries_with_data:
        group['SafeWater_pct'] = group['SafeWater_pct'].interpolate(
            method='linear', limit_direction='both'
        )
    # Egypt and Libya → NaN intentionally kept
    return group
fao_pivot = fao_pivot.groupby('Country', group_keys=False).apply(impute_safewater)
print("✅ SafeWater_pct — linear interpolation applied for Tunisia, Algeria, Morocco")
print("✅ SafeWater_pct — NaN intentionally preserved for Egypt and Libya (FAO data not available)")
```

```python
# ============================================================
# STEP 3 — FoodSupply_variability
# Median by country (volatile — resistant to outliers)
# ============================================================
fao_pivot['FoodSupply_variability'] = fao_pivot.groupby('Country')['FoodSupply_variability'].transform(
    lambda x: x.fillna(x.median())
)
print("✅ FoodSupply_variability — median by country applied")
```

```python
# ============================================================
# STEP 4 — FINAL CHECK
# ============================================================
print("\n=== MISSING VALUE AFTER IMPUTATION ===")
for col in ['Undernourishment_pct', 'SafeWater_pct', 'FoodSupply_variability']:
    print(f"\n{col}:")
    print(fao_pivot.groupby('Country')[col].apply(lambda x: x.isnull().sum()))
```

```python
# ============================================================
# STEP 5 — METHODOLOGICAL DOCUMENTATION
# ============================================================
imputation_log = pd.DataFrame({
    'Variable': [
        'Undernourishment_pct',
        'SafeWater_pct — Tunisia, Algeria, Morocco',
        'SafeWater_pct — Egypt, Libya',
        'FoodSupply_variability'
    ],
    'Method': [
        'Linear interpolation by country',
        'Linear interpolation by country',
        'No imputation — NaN retained',
        'Median by country'
    ],
    'Rationale': [
        'Gradual temporal trend — interpolation preserves trajectory',
        'Gradual temporal trend — interpolation appropriate',
        'Data not available from FAO — absence is informative, imputation would introduce bias',
        'Volatile indicator — median robust to extreme values'
    ]
})

print("\n=== METHODOLOGICAL DOCUMENTATION ===")
imputation_log
```

```python
# --- Identify empty columns ---
empty_columns = wb_data.columns[wb_data.isnull().all()]
print("Empty columns:", empty_columns.tolist())

# --- Identify empty rows ---
empty_rows = wb_data[wb_data.isnull().all(axis=1)]
print("Number of completely empty rows:", len(empty_rows))
```

```python
# Drop rows with index from 120 to 124 - missing values NaN
wb_data = wb_data.drop(wb_data.index[120:125])
wb_data.tail()
```

```python
print(wb_data.isnull().sum())  # missing values
```

```python
'''
In practice, we will:
- Use the score (0–100) for comparisons across countries and years.
- Refer to the confidence interval (CI) and standard error (SI) when we want to highlight reliability and robust analysis:
 If SE is high → the score is less reliable → note this in your conclusions
 If CI is wide → large uncertainty → avoid over‑interpreting comparisons
# Practical rule:
 SE < 0.05 → Score is reliable ✅
 SE > 0.15 → Interpret with caution ⚠️

'''

# Keep only rows where 'Series Name' contains 'score' but NOT 'confidence interval'


wb_data_score = wb_data[
    wb_data['Series Name'].str.contains("score", case=False, na=False) &
    ~wb_data['Series Name'].str.contains("confidence interval", case=False, na=False)
]
wb_data_score.head(10)
```

```python
# Reshape: years as rows instead of columns
wb_data_score = wb_data_score.melt(
    id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"],
    var_name = "Year",
    value_name = "Score"
)

# Extract year : '2000 [YR2000]' → 2000
wb_data_score['Year'] = wb_data_score['Year'].str.extract(r'(\d{4})').astype(int)

wb_data_score.head()
```

```python
wb_data_score = wb_data_score.drop(columns=['Country Code', 'Series Code'], axis=1)
wb_data_score = wb_data_score.rename(columns={'Country Name':'Country', 'Series Name':'Indicator'})
# Replace "Egypt, Arab Rep." with "Egypt" in the Country column
wb_data_score['Country'] = wb_data_score['Country'].replace("Egypt, Arab Rep.", "Egypt")
wb_data_score.head()
```

```python
# renaming the WB “Indicator” column values into short, clear names 

indicator_map = {
    'Government Effectiveness - Governance score (0-100)': 'Gov_Effectiveness',
    'Rule of Law - Governance score (0-100)': 'Rule_of_Law',
    'Control of Corruption - Governance score (0-100)': 'Control_Corruption',
    'Voice and Accountability - Governance score (0-100)': 'Voice_Accountability'
}
wb_data_score['Item_short'] = wb_data_score['Indicator'].map(indicator_map)
wb_data_score = wb_data_score.pivot_table(
    index=['Country', 'Year'],
    columns='Item_short',
    values='Score',
    aggfunc='first'
).reset_index()
wb_data_score.columns.name = None

wb_data_score.head()
```
